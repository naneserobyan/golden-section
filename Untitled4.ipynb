```python
import math
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**4 - 8*x + 10

a = 0
b = 2
eps = 0.5

x_min_plot = -1
x_max_plot = 3

step = 1
while abs(b - a) > eps:
    x1 = a + 0.3819*(b - a)
    x2 = a + 0.6180*(b - a)

    f1 = f(x1)
    f2 = f(x2)

    if f1 < f2:
        b = x2
    else:
        a = x1

    print(f"Step {step}: a={a:.4f}, b={b:.4f}, x1={x1:.4f}, x2={x2:.4f}")


    x_vals = np.linspace(x_min_plot, x_max_plot, 200)
    y_vals = [f(x) for x in x_vals]

    plt.figure()
    plt.plot(x_vals, y_vals, label="f(x)")
    plt.axvline(a, color='green', linestyle='--', label='a')
    plt.axvline(b, color='red', linestyle='--', label='b')
    plt.scatter([x1, x2], [f1, f2], color='orange', label='x1, x2')

    plt.title(f"Քայլ {step}  [a={a:.3f}, b={b:.3f}]")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.grid(True)
    plt.xlim(x_min_plot, x_max_plot)
    plt.ylim(min(y_vals)-1, max(y_vals)+1)
    plt.legend()
    plt.show()

    step += 1

xmin = (a + b) / 2
ymin = f(xmin)
print("\nMinimum found:")
print("xmin =", xmin)
print("f(xmin) =", ymin)
```
